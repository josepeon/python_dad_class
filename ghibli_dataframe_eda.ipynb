#Ghibli Exploratory Data Analysis
Analysis of Studio Ghibli characters and movies using the Ghibli API. DataFrame Generation and analysis work done with pandas.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests, time
```

```python
api = "https://ghibliapi.vercel.app"
def get_data(endpoint):
    response = requests.get(f"{api}/{endpoint}")
    return response.json()
```

```python
films_data = get_data("films")
people_data = get_data("people")
films_df = pd.DataFrame(films_data)
people_df = pd.DataFrame(people_data)
```
